```python
from sklearn.tree import _tree
import numpy as np

def tree_to_code(tree, feature_names):
    """
    Outputs a Python function that represents the decision tree logic.

    Parameters:
    - tree: Trained DecisionTreeRegressor or DecisionTreeClassifier.
    - feature_names: List of feature names used in the model.
    """
    tree_ = tree.tree_
    # Map feature indices to feature names, handle undefined features
    feature_name = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in tree_.feature
    ]
    print("def tree({}):".format(", ".join(feature_names)))

    def recurse(node, depth):
        indent = "  " * depth
        # Check if the node is a split node
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]
            print("{}if {} <= {}:".format(indent, name, threshold))
            recurse(tree_.children_left[node], depth + 1)
            print("{}else:  # if {} > {}".format(indent, name, threshold))
            recurse(tree_.children_right[node], depth + 1)
        else:
            # Leaf node: output the predicted value (regression) or class (classification)
            print("{}return {}".format(indent, np.argmax(tree_.value[node])))
    recurse(0, 1)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import make_regression

# Generate a synthetic regression dataset with 100 samples and 4 features
X, y = make_regression(n_samples=100, n_features=4, noise=0.1)

# Initialize a DecisionTreeRegressor with a maximum depth of 3
tree = DecisionTreeRegressor(max_depth=3)

# Fit the decision tree to the data
tree.fit(X, y)

# Define feature names for readability in the output code
feature_names = ["feature_1", "feature_2", "feature_3", "feature_4"]

# Print out the Python function representing the trained decision tree logic
tree_to_code(tree, feature_names)
# The output will be a Python function that represents the decision tree structure and logic.
```
